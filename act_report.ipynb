# Analysis Report for WeRateDogs Project

After going through the data wrangling procedure that's the gathering, accessing and cleaning,  We made use of the final dataset that's the twitter archives master to derive some key data insights that will be useful and also create a picture for someone reading the dataset.


Overall, the features I analyzed were the source of the tweet, the various dog types we have, favorite count, and the retweets. From the analysis, I got to find out that majority of the tweets were sourced from the iPhone which was around 1353 followed by that the web clients and the TweetDeck. Again, in terms of the number of tweets, the dog type "pupper" had the most mentioned in the various tweets.  It had about 119 followed by doggo with 38. Favorite count was also analyzed with “pupper” topping with a hooping  869685 favorite count followed by doggo with 713377 and “puppo” with 187826 then “floofer” being last with 47795.


Let’s move on to some visualizations in which I analyzed the retweets of these dog types I’ve been talking about in the above insights. 



```python
# Imporing visualizations
# call pacakges
import pandas as pd  
# load dataset
pd_data = pd.read_csv('twitter_archive_master.csv')
# Grouping retweets_count by dogs_type 
visual = pd_data.groupby('dogs_type')['retweet_count'].sum()
# Plotting
visual.plot(xlabel="Dog Type", ylabel="Total Retweets",kind="bar", figsize=(9, 9), title ="Retweets Against Dog types");
```

Visually, it can be seen that the number of favorite count estimated in the insights earlier may consitute the reason why dog type pupper again topping the number of retweets from the figure above. Doggo was the second with around 322814 retweets, followed by puppo with 177842 and lastly floofer with a sum of 7090 retweets.
